```python
import requests
from bs4 import BeautifulSoup
import json

year = 2023

conferences = {
    2024: {
        "name": "The Seventh Annual Conference on Machine Learning and Systems",
        "url": "https://mlsys.org/Conferences/2024/AcceptedPapers"
    },
    2023: {
        "name": "The Sixth Annual Conference on Machine Learning and Systems",
        "url": "https://mlsys.org/Conferences/2023/AcceptedPapers"
    }
}

output_file = "mlsys" + str(year) + ".json"
```

```python
# URL of the page to scrape
url = conferences[year]["url"]

# Send a request to fetch the HTML content
response = requests.get(url)
webpage = response.text

# Parse the HTML content
soup = BeautifulSoup(webpage, 'html.parser')

# Find the table containing the accepted papers
table = soup.find('table')
```

```python
# Initialize a list to store paper details
papers = []

# print(table)

# Extract table rows
for row in table.find_all('tr')[1:]:  # Skipping the header row
    columns = row.find_all('td')
    # print(columns)
    # print("====================================")
    if len(columns) == 3:  # Assuming there are two columns: paper title and authors
        # print(columns[0])

        if columns[0].find_all('a'):
            # print(columns[0].find_all('a')[0].get_text(strip=True))
            paper_title = columns[0].find_all('a')[0].get_text(strip=True)
            paper_authors = columns[0].find_all('i')[0].get_text(strip=True)
            papers.append({'title': paper_title, 'authors': paper_authors})
        elif columns[0].find_all('strong'):
            # print(columns[0].find_all('strong')[0].get_text(strip=True))
            paper_title = columns[0].find_all('strong')[0].get_text(strip=True)
            paper_authors = columns[0].find_all('i')[0].get_text(strip=True)
            papers.append({'title': paper_title, 'authors': paper_authors})

# # Convert list of papers to JSON
# papers_json = json.dumps(papers, indent=4)

# # Save to a JSON file
# with open('accepted_papers.json', 'w') as json_file:
#     json_file.write(papers_json)

# # Print to verify
# print(papers_json)
```

```python
# papers = papers[1:2]
# Transform the authors
conferences_name = conferences[year]["name"]
for entry in papers:
    authors = entry["authors"].split(" \u00b7 ")
    entry["authors"] = [{"first_name": author.split(" ")[0], "last_name": " ".join(author.split(" ")[1:])} for author in authors]
    entry["publication"] = conferences_name
    entry["pubTime"] = year
    entry["pubType"] = 1
```

```python
papers_json = {}
papers_json["info"] = {"name": "Conference on Machine Learning and Systems",
    "short_name": "MLSys",
    "date": str(year)}
papers_json["collection"] = papers

# Convert list of papers to JSON
papers_json = json.dumps(papers_json, indent=4)

# Save to a JSON file
with open(output_file, 'w') as json_file:
    json_file.write(papers_json)

# Print to verify
print(papers_json)
```
